```python
# Building 2 classificator

my_features = ['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'zipcode']

advanced_features = [
    'bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'zipcode',
    'condition', # condition of house
    'grade', # measure of quality of construction
    'waterfront', # waterfront property
    'view', # type of view
    'sqft_above', # square feet above ground
    'sqft_basement', # square feet in basement
    'yr_built', # the year built
    'yr_renovated', # the year renovated
    'lat', 'long', # the lat-long of the parcel
    'sqft_living15', # average sq.ft. of 15 nearest neighbors
    'sqft_lot15', # average lot size of 15 nearest neighbors
]

train_dataset, test_dataset = sf.random_split(fraction=.8, seed=0)
```

```python
simple_model = turi.linear_regression.create(dataset=train_dataset, target='price', features=my_features, validation_set=None)
advanced_model = turi.linear_regression.create(dataset=train_dataset, target='price', features=advanced_features, validation_set=None)

print("Difference rmse:", simple_model.evaluate(test_dataset)['rmse'] - advanced_model.evaluate(test_dataset)['rmse'])

```
